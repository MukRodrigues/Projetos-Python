# Classificador de notas de alunos
Para ajudar professores a dizer se seus alunos estão aprovados, reprovados ou em recuperação de acordo com usas notas.

```python
#classificador 1

'''
dá pra criar um classficador??
'''

def classificador():

  nome=input('Nome do aluno:', )

  média=float(input('Média do aluno:', ))

  if média < 4.1 :
    print('O aluno',nome, 'está reprovado')
  elif média > 6.9:
    print('O aluno',nome, 'está aprovado')
  else:
    print('O aluno',nome, 'está em recuperação')


classificador()
```

```python
#classificador 2

'''
dá pra criar um outro classficador baseado na média das notas das provas??
'''

def classificador2():

  nome=input('Nome do aluno:', )

  nota_1 = float(input('Nota da P1:'), )
  nota_2 = float(input('Nota da P2:'), )
  trabalho = float(input('Nota do trabalho:'), )
  teste = float(input('Nota do teste:'), )


  média= (nota_1 + nota_2)/2 + (trabalho + teste)/10

  if média < 4.1 :
    print('O aluno',nome, 'está reprovado')
  elif média > 6.9:
    print('O aluno',nome, 'está aprovado')
  else:
    print('O aluno',nome, 'está em recuperação')
  return média


classificador2()
```
