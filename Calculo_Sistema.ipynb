### Matrizes do Sistema

```python
A = array([[0.1667, 0.0833, 0      , 0      , 0 ],
[0.0833 , 0.3334 , 0.0833 , 0      , 0 ],
[0      , 0.0833 , 0.3334 , 0.0833 , 0],
[0      , 0      , 0.0833 , 0.3334 , 0.0833],
[0      , 0      , 0      , 0.0833 , 0.1667 ]])

B = array([[2.333 , -1.833 , 0      , 0      , 0  ],
[-1.833 , 4.666 , -1.833 , 0      , 0  ],
[0      , -1.833 , 4.666 , -1.833 , 0],
[0      , 0      , -1.833 , 4.666 , -1.833],
[0      , 0      , 0      , -1.833 , 2.333 ]])

F = array([0.1259, 0.4489,0, -0.4484, -0.1257 ])

c0 = array([0, 0.101321, 0, -0.101321, 0])

k = 0.25

def soma_vet_forca(F,k,n):
  soma = (exp( (n+1)*k   ) + exp( (n)*k))   #F = F_{n+1} + F_{n}
  return F*soma
```

```python
soma_vet_forca(F,k,0)
```

Sistema

$$\left(A+ \frac{k}{2} B \right) c_{n+1} = \left(A- \frac{k}{2} B \right)c_{n} + \frac{k}{2}(F_{n+1} + F_{n} ) $$

```python
k = 0.25
soma = A + k/2 * B
soma
```

```python
subt = A - k/2 * B
subt
```

```python
#lado direito da equação para n = 0
(dot(subt, c0) + k/2* soma_vet_forca(F,k,0)).round(4)
```

```python
#calculando c_1 a partir de c0
def sol_sistema(n):
  c = dot(dot(inv(soma), subt), c0) + k/2* dot(dot(inv(soma), subt), soma_vet_forca(F,k,n))
  return c

c1 = sol_sistema(0).round(3)
c1
```

```python
#lado direito da equação para n = 1
(dot(subt, c1) + k/2* soma_vet_forca(F,k,1)).round(4)
```

```python
#calculando c_2 a partir de c1
def sol_sistema(n):
  c = dot(dot(inv(soma), subt), c1) + k/2* dot(dot(inv(soma), subt), soma_vet_forca(F,k,n))
  return c

c2 = sol_sistema(1).round(4)
c2
```

```python
# L1 <- L1 / a11
A0[0:1] = A0[0:1] /  A0[0:1, 0]      #para acessar os elementos [linha:coluna, elemento]

A0
```

```python
# L2 <- L2 - a21* L1
A0[1:2] = A0[1:2] - A0[1:2, 0] *A0[0:1]
A0[1:2] = A0[1:2] / A0[1:2,1]
A0
```

```python
# L3 <- L3 - a31* L1
A0[2:3] = A0[2:3] - A0[2:3, 0] *A0[0:1]

# L3 <- L3 - a32* L2
A0[2:3] = A0[2:3] - A0[2:3, 1] *A0[1:2]

# L <- L3 / a33
A0[2:3]  = A0[2:3] / A0[2:3, 2]
A0
```

```python
# L4 <- L4 -L3 * a33
A0[3:4] = A0[3:4] - A0[3:4,2]* A0[2:3]   #- A0[2:3, 0] *A0[0:1]

# L4 <- L4 / a44
A0[3:4]  = A0[3:4] / A0[3:4, 3]
A0
```

```python
# L5 <- L5 -L4 * a44
A0[4:5] =A0[4:5] - A0[4:5,3]* A0[3:4]   #- A0[2:3, 0] *A0[0:1]

# L5 <- L5 / a55
A0[4:5] = A0[4:5] / A0[4:5, 4]

print("Logo, após a eliminação de Gauss a matriz aumentada fica:")
print("\n")
A0
```

```python
c5 = A0[4:5,5]
c4 = A0[3:4,5] - A0[3:4,4] *c5
c3 = A0[2:3,5] - A0[2:3,3]*c4
c2 = A0[1:2,5] - A0[1:2,5]*c3
c1 = A0[0:1,5] - A0[0:1,1]*c2

C1 = array([c1[0], c2[0], c3[0], c4[0], c5[0] ]).round(4)   #vetor solução para n=0
C1
```
